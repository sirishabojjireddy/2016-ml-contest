```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import opinion_lexicon
from google.colab import files

# Ensure necessary NLTK resources are available
nltk.download('opinion_lexicon')

# UNOmaha colors (Balanced Theme)
uno_colors = ["#000000", "#BA0C2F", "#FFFFFF"]  # Black, Red, White
colormap = "Greys"  # Using "Greys" for a more neutral visual tone

# Load positive and negative words from NLTK opinion lexicon
positive_words = set(opinion_lexicon.positive())
negative_words = set(opinion_lexicon.negative())

# Function to generate and display a word cloud
def generate_word_cloud(text_list, max_words=200, background_color="#FFFFFF", colormap=colormap):
    # Combine text into a single string
    full_text = " ".join(text_list)

    # Generate word cloud
    wordcloud = WordCloud(width=800, height=400, max_words=max_words,
                          background_color=background_color, colormap=colormap).generate(full_text)

    # Display word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()

# Function to extract neutral words from text
def extract_neutral_words(text):
    words = text.split()  # Tokenize text by spaces
    return " ".join([word.lower() for word in words if word.lower() not in positive_words and word.lower() not in negative_words])  # Keep only neutral words

# Step 1: Upload the Excel file
uploaded = files.upload()  # Manually upload the file

# Step 2: Get the file name dynamically
file_name = list(uploaded.keys())[0]  # Get the uploaded file name

# Step 3: Load Excel file
df = pd.read_excel(file_name)  # Read the uploaded Excel file

# Step 4: Update with actual column name
column_name = "One sentence/ Statement/ Quote  about your experience with ChatGPT Enterprise "  # Replace with the actual column name containing text

if column_name in df.columns:
    df["filtered_text"] = df[column_name].astype(str).apply(extract_neutral_words)  # Apply filtering
    text_list = df["filtered_text"].dropna().tolist()  # Convert to list
    generate_word_cloud(text_list)  # Generate word cloud with UNOmaha colors
else:
    print(f"Error: Column '{column_name}' not found in the Excel file.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import tkinter as tk
from tkinter import filedialog

# Download the VADER lexicon if not already available
nltk.download('vader_lexicon')

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to determine sentiment
def get_sentiment(text):
    if isinstance(text, str):  # Ensure input is a string
        sentiment_score = sia.polarity_scores(text)['compound']
        if sentiment_score >= 0.05:
            return "Positive"
        elif sentiment_score <= -0.05:
            return "Negative"
        else:
            return "Neutral"
    return "Neutral"  # Default if no valid text

# Step 1: Open file dialog to select an Excel file
root = tk.Tk()
root.withdraw()  # Hide the root window
file_path = filedialog.askopenfilename(title="Select an Excel file", filetypes=[("Excel files", "*.xlsx *.xls")])

# Step 2: Load Excel file
if file_path:
    df = pd.read_excel(file_path)  # Read the uploaded Excel file

    # Step 3: Update with actual column name
    column_name = "One sentence/ Statement/ Quote  about your experience with ChatGPT Enterprise "  # Replace with the actual column name containing text

    if column_name in df.columns:
        df["Sentiment"] = df[column_name].astype(str).apply(get_sentiment)  # Apply sentiment analysis

        # Count occurrences of each sentiment
        sentiment_counts = df["Sentiment"].value_counts()

        # Plot the sentiment distribution
        plt.figure(figsize=(8, 5))
        plt.bar(sentiment_counts.index, sentiment_counts.values, color=["green", "red", "gray"])
        plt.xlabel("Sentiment")
        plt.ylabel("Count")
        plt.title("Sentiment Distribution")
        plt.xticks(rotation=0)
        plt.show()

        # Save results to a CSV file
        output_file = "sentiment_results.csv"
        df.to_csv(output_file, index=False)
        print(f"Sentiment analysis results saved as {output_file}.")

    else:
        print(f"Error: Column '{column_name}' not found in the Excel file.")
else:
    print("No file selected.")
```
